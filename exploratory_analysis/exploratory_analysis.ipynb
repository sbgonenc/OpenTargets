```python
## Get location values
def get_loc_values(x:dict):
    try:
        return x.get("location", None), x.get("labelSL", None)
    except Exception as error:
        print(x)
        raise error
```

```python
# Function to convert string to list
import ast
import numpy as np
def convert_to_list(string):
    if string in {None, np.nan}: return None
    try:
        return ast.literal_eval(string)
    except (ValueError, SyntaxError) as e:
        print(f"Error at\n {string}")
    raise e
```

```python
## Get location counts
import json
locations_count = {}
sl_labels_count = {}
for entry in df_targets_simplified["subcellularLocations"]:
    #locs_array = convert_to_list(entry)
    locs_array = entry
    if locs_array == []: continue
    for loc_dict in locs_array:
        loc = loc_dict["location"]
            
        if loc in locations_count:
            locations_count[loc] += 1
        else:
            locations_count[loc] = 1
        ## Some entries do not contain labelSL key
        if "labelSL" not in loc_dict: continue 
        lable = loc_dict["labelSL"]
        if lable in sl_labels_count:
            sl_labels_count[lable] += 1
        else:
            sl_labels_count[lable] = 1
```

```python
def plot_heatmap(column1, column2, cmap="viridis"):
    import seaborn as sns
    import numpy as np

    contingency_table = pd.crosstab(df[column1], df[column2])
    log_contingency_table = np.log2(contingency_table + 1)

    # Create a heatmap using seaborn
    sns.heatmap(log_contingency_table, annot=False, cmap=cmap)
    plt.title(f"{column2} Distribution by {column1} (Log Scale)")
    plt.xlabel(column2)
    plt.ylabel(column1)
    plt.show()
```

```python
import seaborn as sns
import numpy as np

# Create a contingency table showing drugType count for each subcellular location
contingency_table = pd.crosstab(df['subcellular_location'], df['drugType'])
log_contingency_table = np.log2(contingency_table + 1)

# Create a heatmap using seaborn
sns.heatmap(log_contingency_table, annot=False, cmap="viridis")
plt.title("Drug Type Distribution by Subcellular Location (Log Scale)")
plt.xlabel("Drug Type")
plt.ylabel("Subcellular Location")
plt.show()
```
